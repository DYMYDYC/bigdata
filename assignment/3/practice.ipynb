```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
# Instantiate the graph
G1 = nx.Graph()
# add node/edge pairs
G1.add_edges_from([(0, 1),
                   (0, 2),
                   (0, 3),
                   (0, 5),
                   (1, 3),
                   (1, 6),
                   (3, 4),
                   (4, 5),
                   (4, 7),
                   (5, 8),
                   (8, 9),
                   (10,11)])
```

```python
G1.nodes()
```

```python
G1.edges()
```

```python
# draw the network G1
nx.draw_networkx(G1)
```

## Is node 7 reachable from node 6?

```python
nx.node_connectivity(G1, 6, 7)
```

## What is the shortest path between node 6 and 7?

```python
print(nx.shortest_path(G1,source=6,target=7))
```

## Is node 11 reachable from node 6?

```python
nx.node_connectivity(G1, 6, 11)
```

## What are the paths from node 2 to node 7?

```python
for path in nx.all_simple_paths(G1, source=2, target=7):
    print(path)
```

## What are the paths from node 6 to node 9?

```python
for path in nx.all_simple_paths(G1, source=6, target=9):
    print(path)
```

## What is the shortest path between node 6 and 9?

```python
print(nx.shortest_path(G1,source=6,target=9))
```

```python
G1.remove_edge(10,11)
G1.remove_node(10)
G1.remove_node(11)
```

```python
nx.draw_networkx(G1)
```

## What is the minimum edges that disconnects the graph?

```python
len(nx.minimum_edge_cut(G1))
```
